```python
#HANDLING OUTLIERS

```

```python
#WE ARE GOING TO FIND Q1,Q2,Q3 AND IQR (INTER QUARTILE RANGE)

Q1=np.quantile(data["price"],0.25)
Q2=np.quantile(data["price"],0.50)
Q3=np.quantile(data["price"],0.75)

IQR=Q3-Q1

print("Q1: ",Q1)
print("Q2: ",Q2)
print("Q3: ",Q3)
print("IQR: ",IQR)
```

```python
#UPPER WHISKER AND LOWER WHISKER

UW=Q3+(1.5*IQR)
LW=Q1-(1.5*IQR)

print("LW: ",LW)
print("UW: ",UW)

#ANY VALUES LOWER THAN LOWER WHISKER AND MORE THAN UPPER WHISKER IS CONSIDERED AS OUTLIERS!!!
```

```python
#REMOVING ALL OUTLIERS WE ARE GOING TO PLOT  MAKERS WITH PRICE

plt.figure(figsize=(24,14))
sns.boxplot(data=data,x="price",y="make",palette="mako")
plt.show()
```

```python
#REMOVING ALL OUTLIERS WITH LOOP

for maker in data["make"].unique():  #FOR LOOP WILL RUN ON MAKER COLUMN OF EACH MAKER AS WE GIVEN UNIQUE
    price=data[data["make"]==maker]["price"] #ALL THE MAKER PRICE DATA WILL BE STORE IN PRICE VARIABLE WITH 
    Q1=np.quantile(price,0.25) #finding Q1
    Q3=np.quantile(price,0.75) #finding Q3
    IQR=Q3-Q1 #FINDIING IQR FOR UPPER WHISKER AND LOWER WHISKER
    UW=Q3+(1.5*IQR) #UPPER WHISKER
    LW=Q1-(1.5*IQR) #LOWER WHISKER
    
    i=price[price>UW].index #ALL THE DATA OF PRICE WILL GET STORED IN i WITH INDEX NUMBER IF PRICE>UW
    if len(i)!=0: #i list length should not have 0 elments. It will store all row(index) number which is >UW
        data.drop(i,inplace=True)
        
   
```

```python
#IN THIS OUTPUT WE REMOVED ALL THE OUTLIERS

plt.figure(figsize=(24,14))
sns.boxplot(data=data,x="price",y="make",palette="rocket")
plt.show()
```
